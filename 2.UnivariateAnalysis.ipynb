```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari file Excel (ganti 'E Commerce Dataset.xlsx' dengan nama file sebenarnya)
data = pd.read_excel('E Commerce Dataset.xlsx')

# Menampilkan 5 baris pertama data
print(data.head())

# Visualisasi distribusi data numerik bergantung pada status Churn
plt.figure(figsize=(12, 6))
for column in data.select_dtypes(include='number').columns:
    sns.boxplot(data=data, x='Churn', y=column)  # Gunakan boxplot untuk variabel numerik
    plt.title(f'{column} berdasarkan Churn')
    plt.xticks(rotation=45)
    plt.show()

# Visualisasi distribusi data kategorial bergantung pada status Churn
for column in data.select_dtypes(include='object').columns:
    if column != 'Churn':  # Hindari membuat plot untuk kolom 'Churn' sendiri
        plt.figure(figsize=(10, 6))
        sns.countplot(data=data, x=column, hue='Churn')
        plt.title(f'{column} berdasarkan Churn')
        plt.xticks(rotation=45)
        plt.show()
```
